# WebScrapping

```python
players = []
rating = []
pts = []
r1 = []
r2 = []
r3 = []
r4 = []
r5 = []
r6 = []
r7 = []
r8 = []
r9 = []
r10 = []
r11 = []

def addToList(row_values):
    for i in range(len(row_values)):
        if i == 1:
            players.append(row_values[i])
        if i == 3:
            rating.append(row_values[i])
        if i == 15:
            pts.append(row_values[i])
        if i == 4:
            r1.append(row_values[i])
        if i == 5:
            r2.append(row_values[i])
        if i == 6:
            r3.append(row_values[i])
        if i == 7:
            r4.append(row_values[i])
        if i == 8:
            r5.append(row_values[i])
        if i == 9:
            r6.append(row_values[i])
        if i == 10:
            r7.append(row_values[i])
        if i == 11:
            r8.append(row_values[i])
        if i == 12:
            r9.append(row_values[i])
        if i == 13:
            r10.append(row_values[i])
        if i == 14:
            r11.append(row_values[i])
        
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Create a Chrome webdriver
driver = webdriver.Chrome()

# Navigate to the webpage
driver.get("https://www.chess.com/events/2023-fide-world-youth-rapid-championships-u18/results")

# Wait for the table element to be present
table_element = WebDriverWait(driver, 3).until(
    EC.presence_of_element_located((By.CLASS_NAME, "crosstable-swiss-component"))
)

# Find all rows in the table
rows = table_element.find_elements(By.TAG_NAME, "tr")

# Loop through all rows
for row_index, row in enumerate(rows):
    # Extract values from each cell in the row
    row_values = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
    addToList(row_values)
    # Print the values of the row
#     print(f"Values of Row {row_index + 1}: {row_values}")

# Close the browser
driver.quit()
```

```python
import pandas as pd

data = {
    "players": players,
    "rating": rating,
    "pts": pts,
    "r1": r1,
    "r2": r2,
    "r3": r3,
    "r4": r4,
    "r5": r5,
    "r6": r6,
    "r7": r7,
    "r8": r8,
    "r9": r9,
    "r10": r10,
    "r11": r11,
}
```

```python
df = pd.DataFrame(data)
df
```
